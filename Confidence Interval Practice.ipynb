```python
df_men = df[df['Gender']=='Male']
df_men.head()
```

```python
df_sales_by_country_size = df_men.groupby(['Size (US)','Country']).agg({'Country':'count'}).unstack()
df_sales_by_country_size
```

```python
df_sales_by_country_size2 = df_women.groupby(['Size (US)','Country']).agg({'Country':'count'}).unstack()
df_sales_by_country_size2
```

```python
df_men_2016 = df_men[df_men['Date'].apply(lambda x:x.year==2016)]
df_men_2016['Country'].value_counts()
```

```python
df_men_2016_usa = df_men_2016[df_men_2016['Country']=='United States']
df_men_2016_usa.head()
```

```python
df_men_size_month_usa = df_men_2016_usa.groupby(['Size (US)',df_men_2016_usa['Date'].dt.strftime('%B')]).agg({'Size (US)':'count'}).unstack()
df_men_size_month_usa
```

```python
df_men_size_month_usa.fillna(0,inplace=True)
```

```python
df_men_size_month_usa.columns.name = 'Months2016'
df_men_size_month_usa
```

```python
df_men_size_month_usa.sum()
```

```python
def calculate_mean(row):
    return row[1:-4].mean()
```

```python
df_men_size_month_usa['Mean'] = df_men_size_month_usa.apply(calculate_mean,axis=1)
df_men_size_month_usa
```

```python
from scipy import stats
df_men_size_month_usa['Standard Error'] = df_men_size_month_usa.apply(stats.sem,axis=1)
df_men_size_month_usa
```

```python
def calculate_confidence_interval(row):
    return stats.t.interval(0.95,11,loc = row[1:-2].mean(),scale = row['Standard Error'])
```

```python
df_men_size_month_usa['ME'] = df_men_size_month_usa['Standard Error'].apply(lambda x:x*2.20)
df_men_size_month_usa
```

```python
df_men_size_month_usa['95% CI'] = df_men_size_month_usa.apply(calculate_confidence_interval,axis=1)
df_men_size_month_usa
```

```python
def inventory_prediction(CI):
    return np.round(CI[1])[0]
```

```python
df_men_size_month_usa['Prediction of Number of Pairs'] = df_men_size_month_usa['95% CI'].apply(inventory_prediction)
```

```python
df_size = pd.read_excel("Practical example Confidence intervals.xlsx",sheet_name="Shoe size conversion table")
df_size.head(15)
```

```python
df_size.columns = df_size.columns.str.strip()
```

```python
df_size = df_size.iloc[4:,df_size.columns.notnull()]
df_size.head()
```

```python
df_men_size = df_size.iloc[:,:5]
df_women_size = df_size.iloc[:,5:]
```

```python
df_men_size.columns.name = 'Men Sizes'
df_men_size.head()
```

```python
df_women_size.columns.name = 'Women Sizes'
df_women_size.head()
```

```python
df_women_ger1 = df_women[df_women['Shop']=='GER1']
df_women_ger1_2016 = df_women_ger1[df_women_ger1['Date'].apply(lambda x:x.year==2016)] 
df_women_size_month_ger1 = df_women_ger1_2016.groupby(['Size (US)',df_women_ger1_2016['Date'].dt.strftime('%m')]).agg({'Size (US)':'count'}).unstack()
df_women_size_month_ger1.fillna(0,inplace=True)
df_women_size_month_ger1.columns.name = 'GRE1'
df_women_size_month_ger1
```

```python
df_women_size_month_ger1.mean(axis=1)
```

```python
df_women_size_month_ger1.sum()
```

```python
def calc_mean(row):
    return row[0:12].mean()
```

```python
df_women_size_month_ger1['Mean'] = df_women_size_month_ger1.apply(calc_mean,axis=1)
df_women_size_month_ger1
```

```python
def calc_var(row):
    return row[0:12].var()
```

```python
df_women_size_month_ger1['Sample Variance'] = df_women_size_month_ger1.apply(calc_var,axis=1)
df_women_size_month_ger1
```

```python
df_women_ger2 = df_women[df_women['Shop']=='GER2']
df_women_ger2_2016 = df_women_ger2[df_women_ger2['Date'].apply(lambda x:x.year==2016)] 
df_women_size_month_ger2 = df_women_ger2_2016.groupby(['Size (US)',df_women_ger2_2016['Date'].dt.strftime('%m')]).agg({'Size (US)':'count'}).unstack()
df_women_size_month_ger2.fillna(0,inplace=True)
df_women_size_month_ger2.columns.name = 'GRE2'
df_women_size_month_ger2
```

```python
df_women_size_month_ger2.sum()
```

```python
df_women_size_month_ger2['Mean'] = df_women_size_month_ger2.apply(calc_mean,axis=1)
df_women_size_month_ger2
```

```python
df_women_size_month_ger2['Sample Variance'] = df_women_size_month_ger2.apply(calc_var,axis=1)
df_women_size_month_ger2
```

```python
df_women_size_month_ger1['Mean 2'] = df_women_size_month_ger2['Mean']
df_women_size_month_ger1['Sample Variance 2'] = df_women_size_month_ger2['Sample Variance']
df_women_size_month_ger1
```

```python
def calc_pooled_var(row):
    return ((11*row.loc['Sample Variance'])+(11*row.loc['Sample Variance 2']))/22
```

```python
df_women_size_month_ger1['Pooled Variance'] = df_women_size_month_ger1.apply(calc_pooled_var,axis=1)
df_women_size_month_ger1
```

```python
def calculate_ME(row):
    return 2.07 * np.sqrt((row.loc['Pooled Variance']/12)+(row.loc['Pooled Variance']/12)) 
```

```python
df_women_size_month_ger1['Margin of Error'] = df_women_size_month_ger1.apply(calculate_ME,axis=1)
df_women_size_month_ger1
```

```python
def calculate_CI(row):
    return ([(row.loc['Mean'] - row.loc['Mean 2'])-row.loc['Margin of Error'],(row.loc['Mean'] - row.loc['Mean 2'])+row.loc['Margin of Error']])
```

```python
df_women_size_month_ger1['95% CI'] = df_women_size_month_ger1.apply(calculate_CI,axis=1)
df_women_size_month_ger1
```
